# Seaborn 라이브러리 - 고급 그래프 도구 

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 타이타닉 데이터셋
titanic = sns.load_dataset('titanic')
titanic
```

```python
titanic.info()
```

```python
# countplot() : 각 카테고리별 데이터가 얼만큼 있는지를 알 수 있는 플롯
#             : 데이터프레임에서만 사용가능
sns.countplot(x='class', data=titanic)
plt.show()
```

```python
# 산점도 : regplot() 
#       : 서로 다른 2개의 연속 변수 사이의 산점도를 그리고 선형회귀분석에 의한 회귀선도 같이 구할 수 있는 그래프
```

```python
sns.set_style('darkgrid')

# 그래프 객체 생성 
fig = plt.figure(figsize=(15, 5))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

# 그래프 그리기 : 선형회귀선 표시
sns.regplot(x = 'age', y = 'fare', data=titanic, ax=ax1, line_kws={'color':'r', 'alpha':0.4, 'lw':3})

# 그래프 그리기 : 선형회귀선 표시 x
sns.regplot(x = 'age', y= 'fare', data=titanic, ax=ax2, fit_reg=False)

plt.show()
```

```python
# 히트맵 : 2개의 범주형 변수를 각각 x, y축에 놓고 데이터를 매트릭스 형태로 분류
```

```python
sns.set_style('darkgrid')

table = titanic.pivot_table(index=['sex'], columns=['class'], aggfunc='size')

# 히트맵 그리기
sns.heatmap(table, annot=True, fmt='d', cmap='YlGnBu', linewidths=0.5, cbar=True)

plt.show()
```

```python
# 조인트그래프 : 산점도를 기본으로 표시하고 히스토그램까지 동시에 보여줌
# 두 변수의 관계와 데이터가 분산되어 있는 정도를 한 눈에 파악하기 좋음
```

```python
j1 = sns.jointplot(x='fare', y='age', data=titanic)
j2 = sns.jointplot(x='fare', y='age', data=titanic, kind='reg')
j3 = sns.jointplot(x='fare', y='age', data=titanic, kind='hex')
j4 = sns.jointplot(x='fare', y='age', data=titanic, kind='kde')

# 차트 제목 표시
j1.fig.suptitle('titanic fare - scatter', size=13)
j2.fig.suptitle('titanic fare - reg', size=13)
j3.fig.suptitle('titanic fare - hex', size=13)
j4.fig.suptitle('titanic fare - kde', size=13)

plt.show()
```
