```python
# 지도에 마커 표시 
import folium
import pandas as pd
```

```python
# 대학교 리스트 읽어들이기
df = pd.read_excel('./서울지역 대학교 위치.xlsx')
df.head()
```

```python
df.columns = ['name', '위도', '경도']
df.head()
```

```python
# 서울 지도 만들기
seoul_map = folium.Map(location=[37.56, 126.97], tiles='Stamen Terrain', zoom_start=13)
```

```python
# 각 대학교 위치 정보를 Marker로 표시
for name, lat, lng in zip(df.name, df.위도, df.경도):
    #print(name)
    folium.Marker([lat, lng], popup=name).add_to(seoul_map)

seoul_map
```

```python
seoul_map.save('./seoul_college.html')
```

```python
# 마커 옵션 추가
for name, lat, lng in zip(df.name, df.위도, df.경도):    
    folium.CircleMarker([lat, lng], popup=name,
                 radius=10,
                 color='brown',
                 fill=True,
                 fill_color='coral',
                 fill_opacity=0.5
                 ).add_to(seoul_map)

seoul_map
```
